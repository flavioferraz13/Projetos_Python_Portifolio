```python
def preencherCelulas (coluna):
    base_carros[f'{coluna}'] = base_carros[f'{coluna}'].fillna('Nan')
    for item, descricao in enumerate (base_carros[f'{coluna}']):
        if descricao == 'Nan':
            modelo = base_carros['Modelo'].iloc[item]

            for  item2, descricao2 in enumerate (base_carros[f'{coluna}']):
                if descricao2 != 'Nan' and base_carros['Modelo'].iloc[item2] == modelo:
                    base_carros[f'{coluna}'].iloc[item] = descricao2
                    print(str(item2) + ' - ' + descricao2 + '// ' + str(item) + ' - ' +  descricao + ' - ' + modelo)
                    break
```

```python
preencherCelulas('Tipo de veículo')
preencherCelulas('Potência do motor')
preencherCelulas('Combustível')
preencherCelulas('Kit GNV')
preencherCelulas('Câmbio')
preencherCelulas('Direção')
preencherCelulas('Cor')
preencherCelulas('Portas')
```

```python
def limites(coluna):
    q1 = coluna.quantile(0.25) # da o numero que está na posição do 0.25
    q3 = coluna.quantile(0.75) # da o numero que está na posição do 0.25
    amplitude = q3 - q1
    return q1 - 1.5 * amplitude, q3 + 1.5 * amplitude 

def excluir_outliers(df, nome_coluna):
    qtde_linhas = base_carros.shape[0] #pega a quantidade de linhas
    lim_inf, lim_sup = limites(df[nome_coluna])
    df = df.loc[(df[nome_coluna] >= lim_inf) & (df[nome_coluna] <= lim_sup), :]
    lim_removidas = qtde_linhas - df.shape[0]
    
    return df, lim_removidas

def diagrama_caixa(coluna):
    fig, (ax1, ax2) = plt.subplots(1,2)
    fig.set_size_inches(15,5)
    sns.boxplot(x = coluna, ax=ax1)
    ax2.set_xlim(limites(coluna)) #define o limite do eixo
    sns.boxplot(x = coluna, ax=ax2)

def histograma(coluna):
    plt.figure(figsize=(15,5))
    sns.histplot(coluna)
    
def grafico_barra(coluna):
    plt.figure(figsize=(15,5))
    ax = sns.barplot(x = coluna.value_counts().index , y = coluna.value_counts())
    ax.set_xlim(limites(coluna))
```
